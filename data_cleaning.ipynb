```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Step 1: Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Step 2: Rename the column for easier use
df.rename(columns={'emoji used': 'emoji_used'}, inplace=True)

# Step 3: Clean emoji data
def clean_emoji_column(x):
    if pd.isna(x):
        return []
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        x = x.strip()
        if x.startswith('[') and x.endswith(']'):
            try:
                return ast.literal_eval(x)
            except Exception:
                return [x]
        else:
            return [x]
    return [x]

df['emoji_used'] = df['emoji_used'].apply(clean_emoji_column)

# Step 4: Flatten list of emojis
df = df.explode('emoji_used')
df = df[df['emoji_used'].notna()]
df['emoji_used'] = df['emoji_used'].astype(str).str.strip()
df = df[df['emoji_used'] != '']

# Step 5: Count and visualize top emojis
emoji_counts = df['emoji_used'].value_counts().head(10)

plt.rcParams['font.family'] = 'Segoe UI Emoji'  # Windows font for emojis
plt.figure(figsize=(10,6))
sns.barplot(x=emoji_counts.index, y=emoji_counts.values, palette='viridis')
plt.title('Top 10 Emojis Used in Instagram Comments', fontsize=14)
plt.xlabel('Emoji', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0, fontsize=14)
plt.tight_layout()
plt.show()
```
